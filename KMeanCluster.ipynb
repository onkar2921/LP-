```python
df=df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
en=LabelEncoder()
```

```python
df['STATUS']=en.fit_transform(df['STATUS'])
df['PRODUCTLINE']=en.fit_transform(df['PRODUCTLINE'])
df['PRODUCTCODE']=en.fit_transform(df['PRODUCTCODE'])
df['CITY']=en.fit_transform(df['CITY'])
df['COUNTRY']=en.fit_transform(df['COUNTRY'])
df['DEALSIZE']=en.fit_transform(df['DEALSIZE'])
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
In the context of K-Means clustering, WCSS stands for Within-Cluster Sum of Squares. It measures the sum of squared distances between each data point and the centroid of its assigned cluster. WCSS is used as a measure of how tightly the data points are clustered around the centroids and is calculated for each potential number of clusters (k) when determining the optimal number of clusters.
```

```python
wcss=[]
for i in range(1,21):
    km=KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.plot( wcss, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
km=KMeans(n_clusters=4)
```

```python
pred=km.fit_predict(df)
```

```python
df['label']=pred
```
